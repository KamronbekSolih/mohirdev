```python
df.drop(['lot','title_status','vin'],axis=1,inplace=True)
```

```python
# Find the pearson correlations matrix
corr = df.corr(method = 'pearson')
corr
```

```python
# correaltions between age and sex columns
c = np.corrcoef(df['price'],df['year'])
print('Correlations between age and sex\n',c)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8), dpi =500)
sns.heatmap(corr,annot=True,fmt=".1f", linewidth=.1)
plt.show()
```
