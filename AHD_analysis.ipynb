```python
import pandas as pd
```

```python
df = pd.read_csv('Heart.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
df['ChestPain'] = df['ChestPain'].astype('category')
df['ChestPain'] = df['ChestPain'].cat.codes
df
```

```python
df['Thal'] = df.Thal.astype('category')
df['Thal'] = df['Thal'].cat.codes
df
```

```python
df['AHD'] = df.AHD.astype('category')
df['AHD'] = df['AHD'].cat.codes
df
```

```python
df.isnull().sum()
```

```python
 df = df.dropna()
```

```python
df.isnull().sum()
```

```python

X = df.drop(columns='AHD')
```

```python
X
```

```python


y = df.AHD
```

```python


y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 21)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
import numpy as np
np.isnan(X_train_scaled).sum()
```

```python
X_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(random_state=0).fit(X_train_scaled,y_train)
```

```python
model.predict(X_train_scaled)
```

```python
model.score(X_train_scaled,y_train)
```

```python
model.score(X_test_scaled,y_test)
```
