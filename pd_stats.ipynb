```python
color_price = {

}
for color in pd.unique(df['color']):
            color_price[color] = df[df['color'] == color ]['price'].mean()
            print(f"{color}: {df[df['color'] == color ]['price'].mean()}")
```

```python
color_price = pd.Series(color_price)
color_price
```

```python
import matplotlib.pyplot as plt
plt.bar(color_price.index,color_price.values)
plt.xlabel('Categories', fontsize=12)  # Adjust the fontsize as needed
plt.ylabel('Values', fontsize=12)
plt.xticks(rotation=90)
```

```python
hgh = color_price.sort_values(ascending=False)[:10]
```
